```python
import numpy as np
x = np.linspace(-2, 2,100)
y = x ** 2
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-",[0,100,100,0,0,100,50,0,100], [0,0,100,100,0,100,130,100,0], "--g")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
#same thing as above, except broken into pieces
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

```python
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x**2)
plt.savefig("squareFunction.png", transparent=True)
```
