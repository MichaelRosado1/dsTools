```python
x = np.linspace(-1.5, 1.5, 30)
px = .8
py = px**2

plt.plot(x, x**2, px, py, "ro")

plt.annotate("Beautiful point", xy=(px, py), xytext=(px-1.3,py+0.5),
                           color="green", weight="heavy", fontsize=14,
                           arrowprops={"facecolor": "lightgreen"})

plt.show()
```

```python
radius = 1
theta = np.linspace(0, 2*np.pi*radius, 1000)
plt.subplot(111, projection='polar')
plt.plot(theta, np.sin(5*theta), "g-")
plt.plot(theta, .5*np.cos(20*theta), "b-")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=.1)
plt.show()
```

```python
plt.contourf(X,Y, Z, cmap=matplotlib.cm.coolwarm)
plt.colorbar()
plt.show()
```

```python
from numpy.random import rand
x, y, scale = rand(3, 100)
scale = 500 * scale**5
plt.scatter(x, y, s=scale)
plt.show()
```

```python
for color in ['red', 'green', 'blue']:
    n = 100
    x, y = rand(2, n)
    scale = 500 * rand(n) ** 5
    plt.scatter(x, y, s=scale, c=color, alpha=.3, edgecolors='blue')
```

```python
import matplotlib.animation as animation

x = np.linspace(-1, 1, 100)
y = np.sin(x ** 2 * 25)
data = np.array([x, y])

fig = plt.figure()
line, = plt.plot([], [], "r-")
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.plot([-.5, .5], [0, 0], "b-", [0, 0], [-.5, .5], "b-", 0, 0, "ro")
plt.grid(True)

plt.title("animation")

def update_line(num, data, line):
    line.set_data(data[..., :num] + np.random.rand(2, num) / 25)
    return line

line_animation = animation.FuncAnimation(fig, update_line, frames = 50, fargs=(data, line), interval=100)
plt.close()
```
